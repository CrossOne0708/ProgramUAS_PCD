```python
def rekamDataWajah():
    wajahDir = 'datawajah'
    cam = cv2.VideoCapture(3)
    cam.set(3, 640)
    cam.set(4, 480)
    faceDetector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    eyeDetector = cv2.CascadeClassifier('haarcascade_eye.xml')
    faceID = entry2.get()
    nama = entry1.get()
    nim = entry2.get()
    kelas = entry3.get()
    Mata_Kuliah = entry4.get()
    ambilData = 1
    
    # Membuat subdirektori jika belum ada
    sub_dir = os.path.join(wajahDir, nama.lower().replace(" ", "-"))
    if not os.path.exists(sub_dir):
        os.makedirs(sub_dir)
    
    while True:
        retV, frame = cam.read()
        abuabu = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = faceDetector.detectMultiScale(abuabu, 1.3, 5)
        
        for (x, y, w, h) in faces:
            frame = cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 255), 2)
            namaFile = str(nim) + '_' + str(nama) + '_' + str(kelas) + '_' + str(ambilData) +'.jpg'
            file_path = os.path.join(sub_dir, namaFile)
            cv2.imwrite(file_path, frame)
            ambilData += 1
            roiabuabu = abuabu[y:y + h, x:x + w]
            roiwarna = frame[y:y + h, x:x + w]
            eyes = eyeDetector.detectMultiScale(roiabuabu)
            for (xe, ye, we, he) in eyes:
                cv2.rectangle(roiwarna, (xe, ye), (xe + we, ye + he), (0, 255, 255), 1)
        
        cv2.imshow('TEKAN Q UNTUK KELUAR', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):  # jika menekan tombol q akan berhenti
            break
        elif ambilData > 30:
            break
    
    selesai1()
    cam.release()
    cv2.destroyAllWindows()
```

```python
def augmentasiwajah(nama):
    from albumentations import (
    HorizontalFlip, RandomBrightnessContrast, HueSaturationValue, 
    Compose, OneOf, MotionBlur, MedianBlur, Blur)
    
    # Mendefinisikan direktori data dan direktori untuk menyimpan gambar hasil augmentasi
    data_dir = 'datawajah'
    augmented_dir = 'augmentedwajah'

    IMAGE_SIZE = (120, 160)
    SEED = 999

    if not os.path.exists(augmented_dir):
        os.makedirs(augmented_dir)

    # Membuat subdirektori dengan nama yang sesuai
    sub_dir = os.path.join(augmented_dir, nama.lower().replace(' ', '-'))
    if not os.path.exists(sub_dir):
        os.makedirs(sub_dir)

    # Image Augmentation
    augmentation = Compose([
        HorizontalFlip(p=0.5),
        RandomBrightnessContrast(p=0.2),
        HueSaturationValue(p=0.2),
        OneOf([
            MotionBlur(p=0.2),
            MedianBlur(p=0.2),
            Blur(p=0.2),
        ], p=0.2),
    ], p=1)

    # Mendapatkan daftar file gambar dari direktori data
    image_files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith('.jpg')]

    # Mengiterasi setiap file gambar dan melakukan augmentasi
    for image_file in image_files:
        image = cv2.imread(image_file)
        image = cv2.resize(image, IMAGE_SIZE)

        # Apply data augmentation
        augmented = augmentation(image=image)
        augmented_image = augmented['image']

        file_name = os.path.splitext(os.path.basename(image_file))[0]
        output_file = os.path.join(sub_dir, f'{file_name}_augmented.jpg')
        cv2.imwrite(output_file, augmented_image)

    print("Proses augmentasi dan penulisan gambar selesai.")
    print(f"Gambar-gambar hasil augmentasi disimpan di subdirektori: {sub_dir}")
```

```python
def trainingWajah():
    wajahDir = 'datawajah'
    latihDir = 'latihwajah'

    def getImageLabel(path):
        imagePaths = [os.path.join(path, f) for f in os.listdir(path)]
        faceSamples = []
        faceIDs = []
        for imagePath in imagePaths:
            PILimg = Image.open(imagePath).convert('L')
            imgNum = np.array(PILimg, 'uint8')
            faceID = int(os.path.split(imagePath)[-1].split('_')[0])
            faces = faceDetector.detectMultiScale(imgNum)
            for (x, y, w, h) in faces:
                faceSamples.append(imgNum[y:y + h, x:x + w])
                faceIDs.append(faceID)
            return faceSamples, faceIDs

    faceRecognizer = cv2.face.LBPHFaceRecognizer_create()
    faceDetector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faces, IDs = getImageLabel(wajahDir)
    faceRecognizer.train(faces, np.array(IDs))
    # simpan
    faceRecognizer.write(latihDir + '/training.xml')
    selesai2()
```

```python
def markAttendance(name):
    with open("PRESENSI.csv", 'r+') as f:
        namesDatalist = f.readlines()
        namelist = []
        yournim = entry2.get()
        yourclass = entry3.get()
        yourMata_Kuliah = entry4.get()
        for line in namesDatalist:
            entry = line.split(';')
            namelist.append(entry[0])
        if name not in namelist:
            now = datetime.now()
            dtString = now.strftime('%H:%M:%S')
            f.writelines(f'\n{name},{yourclass},{yournim},{dtString}, {yourMata_Kuliah}')
```

```python
def absensiWajah():
    wajahDir = 'datawajah'
    latihDir = 'latihwajah'
    cam = cv2.VideoCapture(0)
    cam.set(3, 640)
    cam.set(4, 480)
    faceDetector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
    faceRecognizer = cv2.face.LBPHFaceRecognizer_create()
    faceRecognizer.read(latihDir + '/training.xml')
    font = cv2.FONT_HERSHEY_SIMPLEX

    #id = 0
    yourname = entry1.get()
    names = []
    names.append(yourname)
    minWidth = 0.1 * cam.get(3)
    minHeight = 0.1 * cam.get(4)

    while True:
        retV, frame = cam.read()
        frame = cv2.flip(frame, 1)
        abuabu = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = faceDetector.detectMultiScale(abuabu, 1.2, 5, minSize=(round(minWidth), round(minHeight)), )
        for (x, y, w, h) in faces:
            frame = cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            id, confidence = faceRecognizer.predict(abuabu[y:y+h,x:x+w])
            if confidence < 100:
                id = names[0]
                confidence = "  {0}%".format(round(150 - confidence))
            elif confidence < 50:
                id = names[0]
                confidence = "  {0}%".format(round(170 - confidence))
            elif confidence > 70:
                id = "Tidak Diketahui"
                confidence = "  {0}%".format(round(150 - confidence))

            cv2.putText(frame, str(id), (x + 5, y - 5), font, 1, (255, 255, 255), 2)
            cv2.putText(frame, str(confidence), (x + 5, y + h + 25), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

        cv2.imshow('TEKAN Q UNTUK KELUAR', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):  # jika menekan tombol q akan berhenti
            break
    markAttendance(id)
    selesai3()
    cam.release()
    cv2.destroyAllWindows()
```
